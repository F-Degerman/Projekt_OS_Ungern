```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

atev = pd.read_csv("../athlete_events.csv")
reg = pd.read_csv("../noc_regions.csv")

df_os = pd.merge(atev, reg, on= "NOC", how= "left")

display(atev.head())
display(reg.head())
display(df_os.head())
```

```python
#a) Hur många länder är med i datan ?
print(f"Antal länder: {len(df_os["NOC"].unique())}")

#b) Vilka länder är med?
print(f"Länder: {df_os["NOC"].unique()}")

#c)Vilka sporter är med?
print(f"Sporter: {df_os["Sport"].unique()}")

#d) Vilka medaljtyper finns?
print(f"Medaljer: {df_os["Medal"].unique()}")

#e) Statistik för ålder: Median , min, max, standardavvikelse
display(df_os["Age"].describe())

#f) Utforska egna frågor
print(f"Tidsinternall: {df_os["Year"].min()} - {df_os["Year"].max()}")

#g) Plotta diagram för könsfördelning
plt.figure(dpi=100, figsize=(6,4))
sns.countplot(data= df_os, x= "Sex", hue= "Sex")

#h) Diagram över top 10 länder som fått flest medaljer.

#i) Plotta fler saker som kan vara intressanta
```
