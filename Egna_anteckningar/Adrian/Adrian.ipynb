```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

atev = pd.read_csv(r"C:\Users\adria\Desktop\Python\Projekt_OS_Ungern\dashboard\data\athlete_events.csv")
reg = pd.read_csv(r"C:\Users\adria\Desktop\Python\Projekt_OS_Ungern\dashboard\data\noc_regions.csv")

df_os = pd.merge(atev, reg, on= "NOC", how= "left")

display(atev.head())
display(reg.head())
display(df_os.head())
```

```python
#Uppgift 1
import hashlib as hl

hun = df_os[df_os["NOC"] == "HUN"]

nameHash = hun["Name"].apply(lambda x: hl.sha256(x.encode()).hexdigest())

hun.insert(1, "nameHash", nameHash)
hun.drop(axis=1, columns="Name", inplace=True)

hun.head()
```

```python
#Total medals by sport
medals = hun.groupby(["Sport", "Medal"]).size().unstack(fill_value=0)
medals["Total"] = medals[["Gold", "Silver", "Bronze"]].sum(axis=1)
medals.sort_values(by= "Total", ascending=False, inplace=True)

uniMedDf = hun.drop_duplicates(subset=["Sport", "Event", "Medal"])
uniqueMedals = uniMedDf.groupby(["Sport", "Medal"]).size().unstack(fill_value=0)
uniqueMedals["Total"] = uniqueMedals[["Gold", "Silver", "Bronze"]].sum(axis=1)
uniqueMedals.sort_values(by= "Total", ascending=False, inplace=True)

top10Sports = uniqueMedals[:10].reset_index()

fig = px.bar(top10Sports, x= "Total", y= "Sport", title="Unique medals per sport", orientation= "h", color= "Total")
fig.update_layout(yaxis=dict(autorange="reversed"))
```

```python
#Medals by sport

medal_count = uniMedDf.groupby(["Sport", "Medal"]).size().reset_index(name="Count")
top_Sports = (
    medal_count.groupby(["Sport"])["Count"].sum()
    .sort_values(ascending=False)[:10]
    .index
)
top10Medals = medal_count[medal_count["Sport"].isin(top_Sports)]

fig = px.bar(
    top10Medals,
    x= "Sport",
    y= "Count",
    color= "Medal",
    barmode= "group",
    title= "Medals by sport Hungary",
    labels= { "Count" : "Number of medals", "Sport" : "Sport"},
    color_discrete_map={'Gold': '#FFD700', 'Silver': '#C0C0C0', 'Bronze': '#CD7F32'}
)
fig.show()

```

```python
#Medals by year
fig = px.bar(
    top10Medals,
    x= "Year",
    y= "Count",
    color= "Medal",
    barmode= "group",
    title= "Medals by Year",
    labels= { "Count" : "Number of medals", "Sport" : "Sport"},
    color_discrete_map={'Gold': '#FFD700', 'Silver': '#C0C0C0', 'Bronze': '#CD7F32'}
)
fig.show()
```

```python
#dela in efter vintersport före 1924

df_modSeasons = df_os.copy() 

DfSeas = df_os[df_os["Year"] > 1924][["Sport", "Season"]]
summerSports = DfSeas[DfSeas["Season"] == "Summer"]["Sport"].unique()
winterSports = DfSeas[DfSeas["Season"] == "Winter"]["Sport"].unique()
df_divByseason = df_os[df_os["Year"] < 1924]

def assign_season(sport):
    if sport in winterSports:
        return "Winter"
    elif sport in summerSports:
        return "Summer"
    else:
        return "Summer"
    
mask = df_modSeasons["Year"] < 1924
df_modSeasons.loc[mask, "Season"] = df_modSeasons.loc[mask, "Sport"].apply(assign_season)

#df_modSeasons.to_csv("Incl_Seasons_pre1924.csv", index=False)


```

```python
medalsbySeason = df_modSeasons[df_modSeasons["NOC"] == "HUN"].groupby(["Year", "Season", "Medal"]).size().reset_index(name="Count")

totPerYear = medalsbySeason.groupby(["Year", "Season"])["Count"].sum().reset_index()

fig = px.bar(totPerYear, x="Year", y="Count", color="Season", barmode="group", title="Hungary, total medals per year")
fig.show()
```
